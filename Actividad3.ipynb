```python
#Leemos el archivo con la función
data= pd.read_csv('Ventas_totales_sin_nulos.csv')
data
```

```python
valores_nulos=data.isnull().sum()
valores_nulos
```

# Columna 1:ventas_precios_corrientes

```python
#Diagrama de caja o bigote
fig = plt.figure(figsize = (5, 3))
plt.boxplot (data["ventas_precios_corrientes"])
plt.title("Outliers de Ventas Precios corrientes")
plt.show() #dibujamos el diagrama
```

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=data["ventas_precios_corrientes"]

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos outliers
outliers_corrientes=data[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_corrientes
```

```python
#Obtenemos datos limpios
data_clean_ventas_corrientes=data[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data_clean_ventas_corrientes
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize = (5, 4))
plt.boxplot (data_clean_ventas_corrientes["ventas_precios_corrientes"])
plt.title("Outliers de Precio corrientes")
plt. show() #dibujamos el diagrama
```
