# HOJA 2020

```python
#Carga desde un archivo .csv sin indice
data= pd.read_excel('Gastos y costos 20-23.xlsx',sheet_name='2020',skiprows=5)
data
```

```python
#Verificamos informaci√≥n del DataFrame
data.info()
```

```python
#Columnas
cuanti=data.iloc[ : , [8,9,10,11,12,13,14,18]]
cuali=data.iloc[ : , [0,1,2,3,4,5,6,7,15,16,17]]
cuali
```

```python
#Corroboramos valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores nulos del dataframe con "bfill"
data1 =data.fillna(method="bfill")
data1 =data1.fillna(method="ffill")
data1
```

```python
#Corroboramos valores nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
cuanti.plot(kind='box', vert=False)
plt.title("Valores At√≠picos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
#M√©todo aplicando desviaci√≥n estandar. Encuentro los valores extremos
y=cuanti
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers del Dataframe
outliers= cuanti[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
data2= cuanti[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data2
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores at√≠picos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=data2.copy()
data_clean=data_clean.fillna(round(data2.mean(),1))
data_clean
```

```python
#Corroboramos valores nulos
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python
Datos_limpios = pd.concat([cuali, data_clean], axis=1)
Datos_limpios
```

```python
#Convertir DataFrame a CSV
Datos_limpios.to_csv("Datos_limpios_2020.csv")
```

# HOJA 2022

```python
#Carga desde un archivo .csv sin indice
data= pd.read_excel('Gastos y costos 20-23.xlsx',sheet_name='2022',skiprows=3)
data
```

```python
#Verificamos informaci√≥n del DataFrame
data.info()
```

```python
#Eliminar columnas innecesarias
#axis = 1 para columnas
#axis = 0 para filas
data= data.drop(['Otros'], axis=1)
data
```

```python
#Columnas
cuanti=data.iloc[ : , [8,9,10,11,12,13,14]]
cuali=data.iloc[ : , [0,1,2,3,4,5,6,7,15,16,17]]
cuali
```

```python
#Corroboramos valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores nulos del dataframe con "bfill"
data1 =data.fillna(method="bfill")
data1 =data1.fillna(method="ffill")
data1
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
cuanti.plot(kind='box', vert=False)
plt.title("Valores At√≠picos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
#M√©todo aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=cuanti

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers del Dataframe
outliers_iqr= cuanti[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios del Dataframe
data1_iqr= cuanti[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data1_iqr
```

```python
#Corroboramos valores nulos
valores_nulos=data1_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores at√≠picos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data2_iqr=data1_iqr.copy()
data2_iqr=data2_iqr.fillna(round(data1_iqr.mean(),1))
data2_iqr
```

```python
#Corroboramos valores nulos
valores_nulos=data2_iqr.isnull().sum()
valores_nulos
```

```python
Datos_limpios = pd.concat([cuali, data2_iqr], axis=1)
Datos_limpios
```

```python
#Convertir DataFrame a CSV
Datos_limpios.to_csv("Datos_limpios_2022.csv")
```
